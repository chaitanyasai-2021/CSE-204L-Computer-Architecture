```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Instagram data.csv')
df.head
```

```python
# Create friendship data from the dataset
friendship_data = df[['Give your username', 'Number of friends in Instagram']].rename(columns={'Give your username': 'User1', 'Number of friends in Instagram': 'User2'})
friendship_data
```

```python
# Create a graph from the friendship data
G = nx.from_pandas_edgelist(friendship_data, 'User1', 'User2')
```

```python
# Visualize the social network graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)  # Position nodes using the spring layout algorithm
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='pink', font_size=8, font_weight='bold')
plt.title('Social Network Graph')
plt.show()
```

## Write a python program to extend the above exercise to discover the communities in the graph, partition the graph and extracting the neighbor hood properties of the graphs.

```python
# Create friendship data from the dataset
friendship_data = df[['Give your username', 'Number of friends in Instagram']].rename(columns={'Give your username': 'User1', 'Number of friends in Instagram': 'User2'})
```

```python
# Create a graph from the friendship data
G = nx.from_pandas_edgelist(friendship_data, 'User1', 'User2')
```

```python
# Convert all node labels to strings
G = nx.relabel.convert_node_labels_to_integers(G, first_label=0, ordering='default', label_attribute=None)

# Discover communities in the graph using Louvain method
communities = nx.algorithms.community.greedy_modularity_communities(G)
```

```python
# Plot the communities
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(G)  # Position nodes using the spring layout algorithm
colors = range(len(communities))
for i, community in enumerate(communities):
    nx.draw_networkx_nodes(G, pos, nodelist=community, node_color=[colors[i]]*len(community), node_size=3000, cmap=plt.cm.Set1, label=f'Community {i+1}')
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.title('Communities in Social Network Graph')
plt.legend()
plt.show()
```

```python
# Partition the graph
partition = nx.algorithms.community.label_propagation.label_propagation_communities(G)
```

```python
# Extract neighborhood properties of the graph
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

# Display neighborhood properties
print("Degree Centrality:", degree_centrality)
print("Closeness Centrality:", closeness_centrality)
print("Betweenness Centrality:", betweenness_centrality)
```
